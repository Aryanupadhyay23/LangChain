### HTMLHeaderTextSplitter

```python
from langchain_text_splitters import HTMLHeaderTextSplitter
```

```python
html_string = """
<!DOCTYPE html>
<html>
<head>
    <title>Artificial Intelligence Blog</title>
</head>
<body>

    <header>
        <h1>Introduction to Artificial Intelligence</h1>
        <p>Published on May 8, 2026</p>
    </header>

    <section>
        <h2>What is Artificial Intelligence?</h2>
        <p>
            Artificial Intelligence (AI) is a branch of computer science
            focused on creating intelligent systems capable of performing
            tasks that normally require human intelligence.
        </p>

        <p>
            AI systems can learn from data, recognize patterns,
            make decisions, and improve over time.
        </p>
    </section>

    <section>
        <h2>Machine Learning</h2>
        <p>
            Machine Learning is a subset of AI that enables systems
            to learn automatically from data without explicit programming.
        </p>

        <ul>
            <li>Supervised Learning</li>
            <li>Unsupervised Learning</li>
            <li>Reinforcement Learning</li>
        </ul>
    </section>

    <section>
        <h2>Deep Learning</h2>
        <p>
            Deep Learning uses neural networks with multiple layers
            to solve complex problems such as image recognition,
            speech processing, and natural language understanding.
        </p>

        <div>
            <h3>Applications</h3>
            <ol>
                <li>Self-driving Cars</li>
                <li>Medical Diagnosis</li>
                <li>Fraud Detection</li>
                <li>Chatbots</li>
            </ol>
        </div>
    </section>

    <section>
        <h2>Generative AI</h2>
        <p>
            Generative AI models like GPT and BERT can generate
            human-like text, summarize documents, answer questions,
            and assist developers in coding tasks.
        </p>
    </section>

    <footer>
        <p>Author: Aryan Upadhyay</p>
        <p>Email: aryan@example.com</p>
    </footer>

</body>
</html>"""
```

```python
headers_to_split_on = [
    ("h1", "Header 1"),
    ("h2", "Header 2"),
    ("h3", "Header 3"),
]

text_splitter = HTMLHeaderTextSplitter(
    headers_to_split_on=headers_to_split_on
)
html_header_splits = text_splitter.split_text(html_string)
```

```python
html_header_splits
```

```python
url = "https://www.aryanupadhyay.com/post/transformer-inference-step-by-step"

headers_to_split_on = [
    ("h1", "Header 1"),
    ("h2", "Header 2"),
    ("h3", "Header 3"),
    ("h4", "Header 4"),
]

html_splitter = HTMLHeaderTextSplitter(headers_to_split_on=headers_to_split_on)
html_header_splits = html_splitter.split_text_from_url(url)
html_header_splits
```

### Recursive JSON Splitter

```python
import json
from langchain_text_splitters import RecursiveJsonSplitter
```

```python
# Load JSON file
with open("sample.json", "r") as f:
    json_data = json.load(f)

# Create JSON splitter
splitter = RecursiveJsonSplitter(
    max_chunk_size=300
)

# Split JSON data
chunks = splitter.split_json(json_data=json_data)

# Print chunks
for i, chunk in enumerate(chunks):
    print(f"\n========== Chunk {i+1} ==========\n")
    print(chunk)
```
