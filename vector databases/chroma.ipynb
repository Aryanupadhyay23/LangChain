#### Chroma

```python
from langchain_chroma import Chroma
from langchain_community.document_loaders import TextLoader
from langchain_community.embeddings import OllamaEmbeddings
from langchain_text_splitters import RecursiveCharacterTextSplitter
```

```python
loader = TextLoader("C:\\Users\\dell\\OneDrive\\Documents\\Desktop\\Langchain\\langchain\\speech.txt")
data = loader.load()
data
```

```python
text_splitter = RecursiveCharacterTextSplitter(chunk_size=500, chunk_overlap=0)
splits = text_splitter.split_documents(data)
```

```python
embeddings = OllamaEmbeddings(model="mxbai-embed-large")
vectordb = Chroma.from_documents(splits, embeddings)
```

```python
query = "What is the speech about?"
docs = vectordb.similarity_search(query)
docs[0].page_content
```

```python
vectordb = Chroma.from_documents(splits, embeddings, persist_directory="./chroma_db" )
```

```python
db2 = Chroma(persist_directory="./chroma_db", embedding_function=embeddings)
db2.similarity_search(query)[0].page_content
```

```python
docs = vectordb.similarity_search_with_score(query)
docs[0]
```

```python
retriever = db2.as_retriever()
retriever.invoke(query)[0].page_content
```
