```python
import os
from dotenv import load_dotenv
load_dotenv()
```

```python
groq_api_key = os.getenv("GROQ_API_KEY")
```

```python
from langchain_groq import ChatGroq

# Initialize Groq LLM model
model = ChatGroq(
    model="llama-3.1-8b-instant",
    groq_api_key=groq_api_key
)
model
```

```python
from langchain_core.messages import HumanMessage
model.invoke([HumanMessage(content = "Hi, My name is Aryan and I am learning GenAI")])
```

```python
from langchain_core.messages import AIMessage
model.invoke([
    HumanMessage(content="Hi, My name is Aryan and I am learning GenAI"),
    AIMessage(content = "Hello Aryan, nice to meet you. Learning GenAI (General Artificial Intelligence) can be a fascinating and challenging subject. GenAI is a broad field that encompasses various areas, including machine learning, natural language processing, computer vision, and more.\n\nAs you dive into GenAI, you'll encounter various concepts, such as:\n\n1. **Machine Learning**: A subset of AI that enables systems to learn from data and improve their performance over time.\n2. **Deep Learning**: A type of machine learning that uses neural networks to analyze complex data, such as images and speech.\n3. **Natural Language Processing (NLP)**: A field that focuses on enabling computers to understand and generate human language.\n4. **Computer Vision**: A field that deals with enabling computers to interpret and understand visual data from images and videos.\n\nWhat specific areas of GenAI are you interested in learning more about, Aryan?"),
    HumanMessage(content="Hey what's my name and what do i do ?")
])
```
