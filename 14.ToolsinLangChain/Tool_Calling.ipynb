```python
from langchain_core.tools import tool, InjectedToolArg
from typing import Annotated
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain_core.messages import HumanMessage, ToolMessage
import requests
import json
```

```python
@tool
def get_conversion_factor(base_currency: str, target_currency: str):
    """
    Fetch conversion rate between two currencies
    """

    url = f"https://v6.exchangerate-api.com/v6/c754eab14ffab33112e380ca/pair/{base_currency}/{target_currency}"

    response = requests.get(url)

    data = response.json()

    return {
        "base_currency": base_currency,
        "target_currency": target_currency,
        "conversion_rate": data["conversion_rate"]
    }
```

```python
@tool
def convert(
    amount: float,
    conversion_rate: Annotated[float, InjectedToolArg]
):
    """
    Convert amount using conversion rate
    """

    return amount * conversion_rate
```

```python
llm = ChatGoogleGenerativeAI(
    model="gemma-4-31b-it",
    api_key=gemini_api_key,
    temperature=0
)

llm_with_tools = llm.bind_tools(
    [get_conversion_factor, convert]
)
```

```python
messages = [
    HumanMessage(
        content="Convert 10 INR to USD"
    )
]
```

```python
ai_message = llm_with_tools.invoke(messages)

messages.append(ai_message)

print(ai_message.tool_calls)
```

```python
tool_call_1 = ai_message.tool_calls[0]

tool_result_1 = get_conversion_factor.invoke(
    tool_call_1["args"]
)

print(tool_result_1)
```

```python
conversion_rate = tool_result_1["conversion_rate"]

print(conversion_rate)
```

```python
messages.append(
    ToolMessage(
        content=json.dumps(tool_result_1),
        tool_call_id=tool_call_1["id"]
    )
)
```

```python
messages
```

```python
ai_message_2 = llm_with_tools.invoke(messages)

messages.append(ai_message_2)

print(ai_message_2.tool_calls)
```

```python
ai_message_2.tool_calls[0]
```

```python
if ai_message_2.tool_calls:
    tool_call_2 = ai_message_2.tool_calls[0]
    tool_call_2["args"]["conversion_rate"] = conversion_rate

    tool_result_2 = convert.invoke(
        tool_call_2["args"]
    )
else:
    print("AI did not generate a tool call for conversion. Manually invoking the convert tool.")
    tool_args_2 = {
        "amount": 10,
        "conversion_rate": conversion_rate
    }
    tool_result_2 = convert.invoke(tool_args_2)

print(tool_result_2)
```

```python
tool_call_2["args"]["conversion_rate"] = conversion_rate
tool_call_2["args"]
```

```python
import uuid

# Check if tool_call_2 was defined in the previous cell
if 'tool_call_2' in locals():
    # tool_call_2 exists, use its ID
    tool_call_id_to_use = tool_call_2["id"]
else:
    # tool_call_2 was not defined, meaning the manual conversion path was taken.
    # Generate a new unique ID for this ToolMessage.
    tool_call_id_to_use = str(uuid.uuid4())

messages.append(
    ToolMessage(
        content=str(tool_result_2),
        tool_call_id=tool_call_id_to_use
    )
)
```

```python
messages
```

```python
final_response = llm_with_tools.invoke(messages)

print(final_response.content)
```
